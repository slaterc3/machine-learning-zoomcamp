```python
import numpy as np 
```

```python
A = np.array([
    [7, 6],
    [17, 9]
])
B = np.array([
    [0, 8],
    [-2, 5]
])
A + B 
```

```python
A = np.array([
    [1, -5, 7],
    [-3, 2, 8]
])

A + (-1 * A)
```

```python
A = np.array([
    [5, -1],
    [0, 11],
    [7, -2]
])
B = np.array([
    [6, 1, 8],
    [-3, 0, 4]
])
A.dot(B)  # matrix multiplication
```

```python
A = np.array([
    [-3, 6],
    [1, 2]
])

# get inverse of A
A_inv = np.linalg.inv(A)
A_inv
```

```python
A = np.array([
    [1, 0, 0],
    [3, 1, -4],
    [-2, -1, 5]
])
np.linalg.inv(A)
```
