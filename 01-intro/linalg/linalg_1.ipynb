```python
import numpy as np 
```

```python
A = np.array([[1,2], [3,2]])
b = np.array([-2,6])

np.linalg.solve(A, b)
```

```python
A = np.array([[2,4], [3,3]])
b = np.array([22,15])

np.linalg.solve(A, b)
```

```python
A = np.array([[1,-3,-6], [0, 1, 2], [-4,12,21]])
b  = np.array([4, -2, -4])
np.linalg.solve(A, b)
```

```python
A = np.array([
    [0, 2, 4],
    [1, 3, 3],
    [2, 7, 6]
])
b = np.array([4, 5, 10])

np.linalg.solve(A, b)
```
