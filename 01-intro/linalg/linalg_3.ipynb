```python
import numpy as np
```

```python
v1 = [1, 1, 1]
v2 = [0, -3, 2]
v3 = [1, 0, 3]

matrix = np.array([v1, v2, v3])
np.linalg.det(matrix)
```

```python
np.linalg.matrix_rank(matrix) == 3
```
